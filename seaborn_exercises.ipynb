```python
sns.boxplot(data = swiss, x = 'is_catholic', y = 'Fertility',)
```

### 4. Using the chipotle dataset from the previous exercise, create a bar chart that shows the 4 most popular items and the revenue produced by each.

```python
chipotle = pd.read_csv('~/codeup-data-science/numpy-pandas-visualization-exercises/chipotle.csv')
```

```python
chipotle
```

```python
popular_items = pd.DataFrame(chipotle.groupby('item_name').item_price.sum().sort_values(ascending = False).head(4))
popular_items
```

```python
sns.barplot(data = popular_items, x = popular_items.index, y = 'item_price')
plt.title('Revenue of Popular Items')
plt.ylabel('Total Revenue (USD)')
plt.xlabel('Menu Item')
```
